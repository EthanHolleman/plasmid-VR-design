# pFC8_9_53_cutters_comparison

```python
from pydna.dseqrecord import Dseqrecord
from pydna.readers import read
from Bio.Restriction import *
```

```python
pFC8 = Dseqrecord(read('../resources/files/genbank/pFC8.gb'))
pFC9 = Dseqrecord(read('../resources/files/genbank/pFC9.gb'))
pFC53tac = Dseqrecord(read('../resources/files/genbank/pFC53tacT1T2.gb'))
```

First find enzymes that could resolve pFC8 and pFC9.

```python
set(pFC9.n_cutters()).difference(set(pFC8.n_cutters()))
```

Checked fridge we don't have any of these.

```python
set(pFC8.n_cutters()).intersection(set(pFC9.twice_cutters()))
```

```python
set(pFC8.n_cutters()).intersection(set(pFC9.no_cutters()))
```

Twice cutters to resolve pFC53 from 9 and 8

```python
set(pFC53tac.twice_cutters()).difference(set(pFC9.twice_cutters()))
```

```python
cutters = pFC8.cutters()

for c in cutters:
    pFC8_cut = pFC8.cut(c)
    pFC9_cut = pFC9.cut(c)
    pfC53tac_cut = pFC53tac.cut(c)
    
    if len(pFC8_cut) != pFC9_cut:
        print(c)
        print(pFC8_cut)
        print(pFC9_cut)
        print(pfC53tac_cut)
        print()
```
